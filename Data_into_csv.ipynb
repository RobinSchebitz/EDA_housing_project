# Import of Data

We want to import the data from the database in the csv file house_sales.csv

## Using psycopg2 and the credentials in our .env file to access PostgreSQL database

```python
import pandas as pd
import psycopg2
import os
from dotenv import load_dotenv

load_dotenv()

DATABASE = os.getenv('DATABASE')
USER_DB = os.getenv('USER_DB')
PASSWORD = os.getenv('PASSWORD')
HOST = os.getenv('HOST')
PORT = os.getenv('PORT')
```

```python
# Establish a connection with psycopg2

conn = psycopg2.connect(
    database=DATABASE,
    user=USER_DB,
    password=PASSWORD,
    host=HOST,
    port=PORT
)
```

```python
# Retrieving the data and importing to a Pandas Dataframe, then closing the connection

query_string = 'SELECT * FROM eda.king_county_house_sales hs LEFT JOIN eda.king_county_house_details hd on hd.id = hs.house_id'
df_psycopg = pd.read_sql(query_string, conn)
conn.close()
```

```python
# Check dataframe

df_psycopg.head()
```

```python
#export the data to a csv-file - be careful it is within the /data folder as it will be ignored by git
df_psycopg.to_csv('data/house_sales.csv',index=False)
```
